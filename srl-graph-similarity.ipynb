```python
import torch
import spacy
import networkx as nx
import pandas as pd
from tqdm import tqdm
from allennlp.predictors import Predictor
from torch_geometric.data import Data, Batch
from transformers import AutoTokenizer, AutoModel
from torch_geometric.nn import GCNConv, global_mean_pool
from scipy import stats
import numpy as np

# Set device
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

# Initialize NLP models
nlp = spacy.load("en_core_web_sm")
srl_predictor = Predictor.from_path("https://storage.googleapis.com/allennlp-public-models/structured-prediction-srl-bert.2020.12.15.tar.gz")
tokenizer = AutoTokenizer.from_pretrained("roberta-large-mnli")
transformer_model = AutoModel.from_pretrained("roberta-large-mnli").to(device)

# 1. Data Loading and Preprocessing
def load_sick_dataset(path):
    df = pd.read_csv(path, sep='\t', 
                    names=['pair_ID', 'sentence_A', 'sentence_B', 
                           'relatedness_score', 'entailment_judgment'],
                    skiprows=1)
    df['relatedness_score'] = (df['relatedness_score'] - 1.0) / (5.0 - 1.0)
    # df = df.head(100)
    return df

# 2. Graph Construction Functions (SRL+SDG)
def get_dependency_graph(sentence):
    """Create dependency graph using spaCy"""
    doc = nlp(sentence)
    dg = nx.DiGraph()
    
    for token in doc:
        dg.add_node(token.i, text=token.text, pos=token.pos_)
        if token.head.i != token.i:  # Avoid root self-reference
            dg.add_edge(token.head.i, token.i, label=token.dep_)
    
    return dg


def construct_srl_sdg_graph(sentence):
    """
    Implement the 4-step SRL+SDG graph construction process from the paper
    
    Args:
        sentence: Input sentence
        
    Returns:
        nx.DiGraph: The constructed SRL+SDG graph
    """
    import matplotlib.pyplot as plt
    import networkx as nx
    
    # Step 1: Get dependency graph
    doc = nlp(sentence)
    dg = nx.DiGraph()
    
    # Build dependency graph
    for token in doc:
        dg.add_node(token.i, text=token.text, pos=token.pos_)
        if token.head.i != token.i:  # Avoid root self-reference
            dg.add_edge(token.head.i, token.i, label=token.dep_)
    
    # Get SRL output
    srl_result = srl_predictor.predict(sentence=sentence)
    
    # Step 1: Create initial SRL graph with predicates as roots and arguments as children
    initial_srl_graph = nx.DiGraph()
    
    # Track argument spans for each predicate
    predicate_arg_spans = {}
    
    for verb_idx, verb in enumerate(srl_result.get("verbs", [])):
        try:
            if "B-V" not in verb["tags"]:
                continue
                
            predicate_idx = verb["tags"].index("B-V")
            predicate_text = srl_result["words"][predicate_idx]
            initial_srl_graph.add_node(predicate_idx, text=predicate_text, type="predicate")
            
            # Find all arguments for this predicate
            arg_spans = {}
            current_arg = None
            start_idx = None
            
            for i, tag in enumerate(verb["tags"]):
                if tag.startswith("B-"):
                    if current_arg and start_idx is not None:
                        arg_spans[current_arg] = (start_idx, i-1)
                    if tag != "B-V":  # Skip the verb itself
                        current_arg = tag[2:]  # Remove "B-" prefix
                        start_idx = i
                elif tag.startswith("I-"):
                    # Continue current argument
                    pass
                elif tag == "O" and current_arg:
                    arg_spans[current_arg] = (start_idx, i-1)
                    current_arg = None
                    start_idx = None
            
            # Add the last argument if it goes to the end
            if current_arg and start_idx is not None:
                arg_spans[current_arg] = (start_idx, len(verb["tags"])-1)
            
            # Store argument spans for this predicate
            predicate_arg_spans[predicate_idx] = arg_spans
            
            # Connect predicate to each argument with labeled edge
            for arg_type, (start, end) in arg_spans.items():
                arg_text = " ".join(srl_result["words"][start:end+1])
                initial_srl_graph.add_node(start, text=arg_text, type="argument", span=(start, end))
                initial_srl_graph.add_edge(predicate_idx, start, label=arg_type)
                
        except Exception as e:
            print(f"Error processing verb: {e}")
            continue
    
    # Step 2: Extract argument subgraphs using dependency grammar
    argument_subgraphs = {}
    
    for pred_idx, arg_spans in predicate_arg_spans.items():
        for arg_type, (start, end) in arg_spans.items():
            # Get all token indices in this argument span
            arg_tokens = list(range(start, end+1))
            
            # Extract subgraph from DG for this argument
            subgraph = nx.DiGraph()
            for u, v, data in dg.edges(data=True):
                if u in arg_tokens and v in arg_tokens:
                    subgraph.add_edge(u, v, label=data['label'])
                    subgraph.nodes[u]['text'] = dg.nodes[u]['text']
                    subgraph.nodes[v]['text'] = dg.nodes[v]['text']
            
            # Store the subgraph
            argument_subgraphs[(pred_idx, arg_type, start)] = subgraph
    
    # Step 3: Build base graph with argument roots
    base_graph = nx.DiGraph()
    


# def construct_srl_sdg_graph(sentence):
#     """Implement the 4-step SRL+SDG graph construction process from the paper"""
#     try:
#         # Step 1: Get dependency graph
#         dg = get_dependency_graph(sentence)
        
#         # Get SRL output
#         srl_result = srl_predictor.predict(sentence=sentence)
        
#         # Step 1: Create initial SRL graph with predicates as roots and arguments as children
#         initial_srl_graph = nx.DiGraph()
        
#         for verb in srl_result.get("verbs", []):
#             try:
#                 if "B-V" not in verb["tags"]:
#                     continue
                    
#                 predicate_idx = verb["tags"].index("B-V")
#                 initial_srl_graph.add_node(predicate_idx, text=verb["verb"], type="predicate")
                
            #     # Find all arguments for this predicate
            #     arg_spans = {}
            #     current_arg = None
            #     start_idx = None
                
            #     for i, tag in enumerate(verb["tags"]):
            #         if tag.startswith("B-"):
            #             if current_arg and start_idx is not None:
            #                 arg_spans[current_arg] = (start_idx, i-1)
            #             if tag != "B-V":  # Skip the verb itself
            #                 current_arg = tag[2:]  # Remove "B-" prefix
            #                 start_idx = i
            #         elif tag == "O" and current_arg:
            #             arg_spans[current_arg] = (start_idx, i-1)
            #             current_arg = None
            #             start_idx = None
                
            #     # Add the last argument if it goes to the end
            #     if current_arg and start_idx is not None:
            #         arg_spans[current_arg] = (start_idx, len(verb["tags"])-1)
                
            #     # Connect predicate to each argument with labeled edge
            #     for arg_type, (start, end) in arg_spans.items():
            #         for i in range(start, end+1):
            #             initial_srl_graph.add_node(i, text=srl_result["words"][i], type="argument")
            #         initial_srl_graph.add_edge(predicate_idx, start, label=arg_type)
            # except Exception as e:
            #     print(f"Error processing verb: {e}")
        #         continue
        
        # # Step 2: Extract argument subgraphs using dependency grammar
        # argument_subgraphs = {}
        
        # for pred_node, arg_node, data in initial_srl_graph.edges(data=True):
        #     arg_role = data['label']
            
        #     # Find all nodes in this argument
        #     arg_nodes = []
        #     for node in initial_srl_graph.nodes():
        #         if initial_srl_graph.nodes[node].get('type') == 'argument':
        #             # Check if this node is part of the current argument
        #             if initial_srl_graph.has_edge(pred_node, node) and initial_srl_graph.edges[pred_node, node]['label'] == arg_role:
        #                 arg_nodes.append(node)
            
        #     # Extract subgraph from DG for this argument
        #     if arg_nodes:
        #         subgraph = nx.DiGraph()
        #         for u, v, data in dg.edges(data=True):
        #             if u in arg_nodes and v in arg_nodes:
        #                 subgraph.add_edge(u, v, label=data['label'])
        #         argument_subgraphs[(pred_node, arg_role)] = subgraph
        
        # # Step 3: Build base graph with argument roots
        # base_graph = nx.DiGraph()
        
        # for (pred_node, arg_role), subgraph in argument_subgraphs.items():
        #     # Find roots of this argument subgraph (nodes with no incoming edges within subgraph)
        #     if not subgraph.nodes():
        #         continue
                
        #     roots = []
        #     for node in subgraph.nodes():
        #         if subgraph.in_degree(node) == 0:
    #                 roots.append(node)
            
    #         # If no roots found, use all nodes
    #         if not roots:
    #             roots = list(subgraph.nodes())
            
    #         # Connect predicate to each root with the semantic role
    #         for root in roots:
    #             base_graph.add_edge(pred_node, root, label=arg_role)
        
    #     # Step 4: Create final SRL+SDG graph
    #     sdg_graph = nx.DiGraph()
        
    #     # First add all DG edges with label 'S'
    #     for u, v, data in dg.edges(data=True):
    #         sdg_graph.add_edge(u, v, label="S")
        
    #     # Then overlay base graph edges, replacing any existing 'S' edges
    #     for u, v, data in base_graph.edges(data=True):
    #         if sdg_graph.has_edge(u, v):
    #             sdg_graph.remove_edge(u, v)
    #         sdg_graph.add_edge(u, v, label=data['label'])
        
    #     return sdg_graph
    
    # except Exception as e:
    #     print(f"Error in SRL+SDG construction: {e}")
    #     # Return dependency graph as fallback
    #     return get_dependency_graph(sentence)


# 3. Graph Neural Network Architecture
class DGNN(torch.nn.Module):
    def __init__(self, input_dim=1024, hidden_dim=512, num_layers=4):
        super().__init__()
        self.num_layers = num_layers
        
        # Encoder GCN layers
        self.enc_convs = torch.nn.ModuleList()
        self.enc_convs.append(GCNConv(input_dim, hidden_dim))
        for i in range(1, num_layers):
            self.enc_convs.append(GCNConv(hidden_dim, hidden_dim))
        
        # Decoder GCN layers
        self.dec_convs = torch.nn.ModuleList()
        for i in range(num_layers):
            self.dec_convs.append(GCNConv(hidden_dim, hidden_dim))
        
        # Attention mechanism
        self.attention = torch.nn.MultiheadAttention(hidden_dim, 8)
        
        # Output layers
        self.dropout = torch.nn.Dropout(0.2)
        self.fc = torch.nn.Linear(hidden_dim*2, 1)
    
    def forward(self, data_a, data_b):
        """
        Forward pass for the DGNN model
        
        Args:
            data_a: PyG Data object for first sentence
            data_b: PyG Data object for second sentence
            
        Returns:
            torch.Tensor: Similarity score between sentences
        """
        # Process sentence A
        x_a = self.process_graph(data_a)
        
        # Process sentence B
        x_b = self.process_graph(data_b)
        
        # Ensure tensors have batch dimension
        if x_a.dim() == 1:
            x_a = x_a.unsqueeze(0)
        if x_b.dim() == 1:
            x_b = x_b.unsqueeze(0)
        
        # Combine representations
        combined = torch.cat([x_a, x_b], dim=1)
        combined = self.dropout(combined)
        
        # Final prediction
        return torch.sigmoid(self.fc(combined))

    def process_graph(self, data):
        """
        Process a single graph through the GCN layers
        
        Args:
            data: PyG Data object containing the graph
            
        Returns:
            torch.Tensor: Graph representation
        """
        x, edge_index = data.x, data.edge_index
        
        # Encoder
        enc_features = []
        for i in range(self.num_layers):
            x = self.enc_convs[i](x, edge_index)
            x = torch.relu(x)
            enc_features.append(x)
        
        # Decoder with skip connections
        for i in range(self.num_layers):
            x = x + enc_features[self.num_layers-i-1]  # Skip connection
            x = self.dec_convs[i](x, edge_index)
            x = torch.relu(x)
        
        # Apply attention
        x_reshaped = x.unsqueeze(0)
        x_attn, _ = self.attention(x_reshaped, x_reshaped, x_reshaped)
        x = x_attn.squeeze(0)
        
        # Global pooling
        if hasattr(data, 'batch') and data.batch is not None:
            pooled = global_mean_pool(x, data.batch)
        else:
            pooled = x.mean(dim=0)
    
        # Ensure output has proper dimensions
        if pooled.dim() == 1:
            return pooled.unsqueeze(0)
        return pooled

# class DGNN(torch.nn.Module):
#     def __init__(self, input_dim=1024, hidden_dim=512):
#         super().__init__()
#         self.conv1 = GCNConv(input_dim, hidden_dim)
#         self.conv2 = GCNConv(hidden_dim, hidden_dim)
#         self.fc = torch.nn.Linear(hidden_dim*2, 1)
        
#     def forward(self, data_a, data_b):  # <-- This must be properly indented
#         # Process Sentence A
#         x_a = self.process_graph(data_a)
        
#         # Process Sentence B
#         x_b = self.process_graph(data_b)
        
#         # Combine representations
#         combined = torch.cat([x_a, x_b], dim=1)
#         return torch.sigmoid(self.fc(combined))
        
#     def process_graph(self, data):
#         x, edge_index = data.x, data.edge_index
        
#         # Encoder
#         enc_features = []
#         for i in range(self.num_layers):
#             x = self.enc_convs[i](x, edge_index)
#             x = torch.relu(x)
#             enc_features.append(x)
        
#         # Decoder with skip connections
#         for i in range(self.num_layers):
#             x = x + enc_features[self.num_layers-i-1]  # Skip connection
#             x = self.dec_convs[i](x, edge_index)
#             x = torch.relu(x)
        
#         # Apply attention
#         x_reshaped = x.unsqueeze(0)
#         x_attn, _ = self.attention(x_reshaped, x_reshaped, x_reshaped)
#         x = x_attn.squeeze(0)
        
#         # Global pooling - MODIFIED TO PRESERVE DIMENSIONS
#         if hasattr(data, 'batch') and data.batch is not None:
#             pooled = global_mean_pool(x, data.batch)
#         else:
#             pooled = x.mean(dim=0)
        
#         # Add batch dimension if missing
#         if pooled.dim() == 1:
#             return pooled.unsqueeze(0)  # Add batch dimension
#         return pooled


# def forward(self, data_a, data_b):
#     # Process sentence A
#     x_a = self.process_graph(data_a)
#     if x_a.dim() == 1:
#         x_a = x_a.unsqueeze(0)
    
#     # Process sentence B
#     x_b = self.process_graph(data_b)
#     if x_b.dim() == 1:
#         x_b = x_b.unsqueeze(0)
    
#     # Combine representations - now both tensors have at least 2 dimensions
#     combined = torch.cat([x_a, x_b], dim=1)
#     combined = self.dropout(combined)
    
#     # Final prediction
#     return torch.sigmoid(self.fc(combined))

    # def process_graph(self, data):
    #     x, edge_index = data.x, data.edge_index
        
    #     # Encoder
    #     enc_features = []
    #     for i in range(self.num_layers):
    #         x = self.enc_convs[i](x, edge_index)
    #         x = torch.relu(x)
    #         enc_features.append(x)
        
    #     # Decoder with skip connections
    #     for i in range(self.num_layers):
    #         x = x + enc_features[self.num_layers-i-1]  # Skip connection
    #         x = self.dec_convs[i](x, edge_index)
    #         x = torch.relu(x)
        
    #     # Apply attention
    #     x_reshaped = x.unsqueeze(0)
    #     x_attn, _ = self.attention(x_reshaped, x_reshaped, x_reshaped)
    #     x = x_attn.squeeze(0)
        
    #     # Global pooling
    #     if hasattr(data, 'batch') and data.batch is not None:
    #         return global_mean_pool(x, data.batch)
    #     else:
    #         return x.mean(dim=0)


# 4. Data Preparation Pipeline
def prepare_data(sentence, max_length=128):
    # Get transformer embeddings
    inputs = tokenizer(sentence, return_tensors="pt", 
                      padding='max_length', max_length=max_length,
                      truncation=True).to(device)
    
    with torch.no_grad():
        outputs = transformer_model(**inputs)
    embeddings = outputs.last_hidden_state.squeeze(0)
    
    # Build SRL+SDG graph
    try:
        graph = construct_srl_sdg_graph(sentence)
        
        # Convert to PyG format
        node_indices = list(graph.nodes())
        if not node_indices:
            raise ValueError("Empty graph")
            
        edge_list = list(graph.edges())
        if not edge_list:
            # Add self-loops if no edges
            edge_list = [(i, i) for i in node_indices]
            
        edge_index = torch.tensor(edge_list).t().contiguous().to(device)
        
        # Get edge attributes (labels)
        edge_attr = []
        for u, v in edge_list:
            if graph.has_edge(u, v):
                label = graph.edges[u, v]['label']
                # Convert label to numeric value
                label_val = hash(label) % 100  # Simple hash function
                edge_attr.append([label_val])
        
        edge_attr = torch.tensor(edge_attr, dtype=torch.float).to(device)
        
        # Ensure embeddings match graph nodes
        x = embeddings[:len(node_indices)].to(device)
        
        return Data(x=x, edge_index=edge_index, edge_attr=edge_attr)
    
    except Exception as e:
        print(f"Error in prepare_data: {e}")
        # Fallback to simple graph
        x = embeddings[:min(len(embeddings), 32)].to(device)
        edge_index = torch.tensor([[i, i+1] for i in range(len(x)-1)]).t().contiguous().to(device)
        return Data(x=x, edge_index=edge_index)

# 5. Batch Processing
def collate_graphs(batch):
    return Batch.from_data_list(batch)

# 6. Training Loop
def train_model(train_df, epochs=10, batch_size=32):
    model = DGNN().to(device)
    
    # Use AdamW with different learning rates
    optimizer = torch.optim.AdamW([
        {'params': model.parameters()},
        {'params': transformer_model.parameters(), 'lr': 2e-5}
    ], lr=0.001)
    
    criterion = torch.nn.MSELoss()
    
    for epoch in range(epochs):
        model.train()
        total_loss = 0
        
        # Create batches
        batches = []
        current_batch = []
        batch_targets = []
        
        progress_bar = tqdm(train_df.iterrows(), total=len(train_df), desc=f"Epoch {epoch+1}")
        
        for _, row in progress_bar:
            try:
                data_a = prepare_data(row['sentence_A'])
                data_b = prepare_data(row['sentence_B'])
                target = row['relatedness_score']
                
                current_batch.append((data_a, data_b))
                batch_targets.append(target)
                
                if len(current_batch) == batch_size:
                    batches.append((current_batch, batch_targets))
                    current_batch = []
                    batch_targets = []
            except Exception as e:
                print(f"Error processing row: {e}")
                continue
        
        # Add the last batch if it's not empty
        if current_batch:
            batches.append((current_batch, batch_targets))
        
        # Process batches
        batch_progress = tqdm(batches, desc=f"Batch Progress")
        for batch_data, batch_targets in batch_progress:
            optimizer.zero_grad()
            
            batch_preds = []
            for data_a, data_b in batch_data:
                pred = model(data_a, data_b)
                batch_preds.append(pred)
            
            # Stack predictions and targets
            batch_preds = torch.cat(batch_preds).view(-1, 1)
            batch_targets = torch.tensor(batch_targets, dtype=torch.float32).view(-1, 1).to(device)
            
            # Compute loss
            loss = criterion(batch_preds, batch_targets)
            loss.backward()
            optimizer.step()
            
            total_loss += loss.item() * len(batch_data)
            batch_progress.set_postfix(loss=loss.item())
        
        print(f"Epoch {epoch+1}, Avg Loss: {total_loss/len(train_df):.4f}")
    
    return model

# 7. Evaluation
def evaluate_model(model, test_df, batch_size=32):
    model.eval()
    predictions = []
    targets = []
    
    # Create batches
    batches = []
    current_batch = []
    batch_targets = []
    
    with torch.no_grad():
        progress_bar = tqdm(test_df.iterrows(), total=len(test_df), desc="Evaluating")
        
        for _, row in progress_bar:
            try:
                data_a = prepare_data(row['sentence_A'])
                data_b = prepare_data(row['sentence_B'])
                target = row['relatedness_score']
                
                current_batch.append((data_a, data_b))
                batch_targets.append(target)
                
                if len(current_batch) == batch_size:
                    batches.append((current_batch, batch_targets))
                    current_batch = []
                    batch_targets = []
            except Exception as e:
                print(f"Error processing test row: {e}")
                continue
        
        # Add the last batch if it's not empty
        if current_batch:
            batches.append((current_batch, batch_targets))
        
        # Process batches
        for batch_data, batch_targets in tqdm(batches, desc="Batch Evaluation"):
            batch_preds = []
            for data_a, data_b in batch_data:
                pred = model(data_a, data_b)
                batch_preds.append(pred.cpu().item())
            
            predictions.extend(batch_preds)
            targets.extend(batch_targets)
    
    # Calculate correlations with proper error handling
    predictions = np.array(predictions)
    targets = np.array(targets)
    
    # Add small noise to prevent constant arrays
    if np.std(predictions) < 1e-6:
        predictions = predictions + np.random.normal(0, 1e-6, len(predictions))
    
    pearson = stats.pearsonr(predictions, targets)[0] if np.std(predictions) > 0 else 0
    spearman = stats.spearmanr(predictions, targets)[0] if np.unique(predictions).size > 1 else 0
    
    return {
        'pearson': pearson,
        'spearman': spearman
    }

# Main Execution
if __name__ == "__main__":
    # Load dataset
    sick_df = load_sick_dataset('/kaggle/input/sickdata12/SICK_train.txt')
    
    # Check dataset
    print(f"Dataset loaded: {len(sick_df)} pairs")
    print(f"Sample: {sick_df.iloc[0]}")
    
    # Verify score distribution
    print(f"Score range: {sick_df['relatedness_score'].min()} to {sick_df['relatedness_score'].max()}")
    print(f"Unique scores: {sick_df['relatedness_score'].nunique()}")
    
    # Split dataset
    train_df = sick_df.sample(frac=0.8, random_state=42)
    test_df = sick_df.drop(train_df.index)
    
    print(f"Training on {len(train_df)} pairs, testing on {len(test_df)} pairs")
    
    # Train model
    model = train_model(train_df, epochs=2, batch_size=16)
    
    # Evaluate
    metrics = evaluate_model(model, test_df)
    print(f"Pearson: {metrics['pearson']:.4f}, Spearman: {metrics['spearman']:.4f}")
    
    # Save model
    torch.save(model.state_dict(), "srl_sdg_model.pt")
```
